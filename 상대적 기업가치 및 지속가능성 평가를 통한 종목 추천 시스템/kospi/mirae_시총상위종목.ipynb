```python
import pandas as pd
import numpy as np
import FinanceDataReader as fdr
from google.colab import drive
drive.mount('/content/drive')
```

```python
# 한국거래소 KOSPI 시세정보 
df_krx = pd.read_csv('/content/drive/MyDrive/미래에셋/kospi/KRX_전종목시세_0912.csv', sep = ",",encoding = 'CP949', error_bad_lines=False) 



# 한국거래소 KOSPI 종목정보
df_1 = fdr.StockListing('KOSPI')
df_2 = df_1[['Name', 'Sector']]
df_2 = df_2.rename(columns = {"Name" : "종목명", "Sector" : "업종"})
# print(df_2)



# KOSPI 전체종목 데이터와 시세정보 데이터 Outer Join
df_3 = pd.merge(df_krx, df_2, how='outer', on = '종목명')
print(df_3)



# 시가총액 상위 500개 종목 추출
df_3 = df_3.sort_values(by = ['시가총액'], ascending = False)
df_4 = df_3.iloc[0:499] 
df_4 = df_4.reset_index(drop=True, inplace = False)
print(df_4.head())
df_4.to_csv("KOSPI_시가총액_상위500.csv")
```
